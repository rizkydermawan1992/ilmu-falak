### Penentuan Julian Day

Julian Day adalah banyaknya hari yang telah dilalui sejak tanggal 1 Januari 4713 SM pada pertengahan hari (12:00:00 UT)

```python
def julianDay(H, B, T):
    if B < 3:
        B = B+12
        T = T-1

    #cek apakah masuk kategori kalender JULIAN atau GREGORIAN
    if T < 1582:
        B2 = 0
        Kalender = "Julian"
    elif T == 1582 and B<10:
        B2 = 0
        Kalender = "Julian"
    elif T == 1582 and B==10 and H<=4: # 4 Oktober 1582 Berakhirnya Kalender Julian
        B2 = 0
        Kalender = "Julian"
    else:
        T2 = int(T/100)
        B2 = 2-T2+int(T2/4) 
        Kalender = "Gregorian"

    #Kalkulasi Julian Day
    JD = 1720994.5 + int(365.25*T) + int(30.6001*(B+1)) + B2 + H 
    return JD, Kalender
    
```

```python
from math import *

H = float(input("Masukkan Tanggal: "))
B = int(input("Masukkan bulan: "))
T = int(input("Masukkan Tahun: "))

JD, Kalender = julianDay(H, B, T)
print(f'Julian Day: {JD}')
print(f'Kalender: {Kalender}')

```

# Menentukan Hari Berdasarkan Julian Day

```python
def namaHari(JD):
    p = JD + 1.5
    ss = (p/7) - int(p/7) #sisa hari
    ss = ss*7

    nomorHari = round(ss+1)

    if nomorHari == 1:
        Hari = "Ahad"
    elif nomorHari == 2:
        Hari = "Senin"
    elif nomorHari == 3:
        Hari = "Selasa"
    elif nomorHari == 4:
        Hari = "Rabu"
    elif nomorHari == 5:
        Hari = "Kamis"
    elif nomorHari == 6:
        Hari = "Jumat"
    elif nomorHari == 7:
        Hari = "Sabtu"
    else:
        Hari = "Tidak Diketahui"
    
    return Hari
   
```

```python
#fungsi konversi angka ke nama bulan

def bulanMasehi(angkaBulan):
    if angkaBulan == 1:
        namaBulan = "Januari"
    elif angkaBulan == 2:
        namaBulan = "Februari"
    elif angkaBulan == 3:
        namaBulan = "Maret"
    elif angkaBulan == 4:
        namaBulan = "April"
    elif angkaBulan == 5:
        namaBulan = "Mei"
    elif angkaBulan == 6:
        namaBulan = "Juni"
    elif angkaBulan == 7:
        namaBulan = "Juli"
    elif angkaBulan == 8:
        namaBulan = "Agustus"
    elif angkaBulan == 9:
        namaBulan = "September"
    elif angkaBulan == 10:
        namaBulan = "Oktober"
    elif angkaBulan == 11:
        namaBulan = "November"
    elif angkaBulan == 12:
        namaBulan = "Desember"
    else:
        namaBulan = "Tidak Diketahui"
        
    return namaBulan

def bulanHijriah(angkaBulan):
    if angkaBulan == 1:
        namaBulan = "Muharram"
    elif angkaBulan == 2:
        namaBulan = "Shafar"
    elif angkaBulan == 3:
        namaBulan = "Rabiul Awal"
    elif angkaBulan == 4:
        namaBulan = "Rabiul Akhir"
    elif angkaBulan == 5:
        namaBulan = "Jumadil Awal"
    elif angkaBulan == 6:
        namaBulan = "Jumadil Akhir"
    elif angkaBulan == 7:
        namaBulan = "Rajab"
    elif angkaBulan == 8:
        namaBulan = "Sya'ban"
    elif angkaBulan == 9:
        namaBulan = "Ramadhan"
    elif angkaBulan == 10:
        namaBulan = "Syawal"
    elif angkaBulan == 11:
        namaBulan = "Dzulqa'dah"
    elif angkaBulan == 12:
        namaBulan = "Dzulhijjah"
    else:
        namaBulan = "Tidak Diketahui"
        
    return namaBulan
```

### Konversi Hijriah ke Masehi (ALGORITMA JEEN - MEUS)

```python
from math import *

H = int(input("Tanggal Hijriah: "))
B = int(input("Bulan Hijriah: "))
T = int(input("Tahun Hijriah: "))

N = H + int(29.5001*(B-1)+0.99)
# print(f'N = {N}')
Q = int(T/30)
# print(f'Q = {Q}')
R = fmod(T, 30)
# print(f'R = {R}')
A = int((11*R+3)/30)
# print(f'A = {A}')
W = 404*Q+354*R+208+A
# print(f'W = {W}')
Q1 = int(W/1461)
# print(f'Q1 = {Q1}')
Q2 = fmod(W, 1461)
# print(f'Q2 = {Q2}')
G = 621+4*int(7*Q+Q1)
# print(f'G = {G}')
K = int(Q2/365.2422)
# print(f'K = {K}')
E = int(365.2422*K)
# print(f'E = {E}')
J = Q2 - E + N - 1
# print(f'J = {J}')
X = G + K
# print(f'X = {X}')

if fmod(X,4) == 0 and J>366:
    J = J-366
    X = X+1
elif fmod(X,4) == 0  and J>365:
    J = J-365
    X = X+1
JD = int(365.25*(X-1))+1721423+J
# print(f'JD = {JD}')


alpha = int((JD-1867216.25)/36524.25)
# print(f'Alpha = {alpha}')
beta = JD+1+alpha-int(alpha/4)
# print(f'Beta = {beta}')
if beta < 2299161:
    beta = JD

b = beta+1524
# print(f'b = {b}')
c = int((b-122.1)/365.25)
# print(f'c = {c}')
d = int(365.25*c)
# print(f'd = {d}')
e = int((b-d)/30.6001)
# print(f'e = {e}')
HM = b-d-int(30.6001*e)

if e< 14:
    BM = e-1
else:
    BM = e-13

if BM<3:
    TM = c-4715
else:
    TM = c-4716
    
JD, _ = julianDay(HM, BM, TM)
Hari = namaHari(JD)
B = bulanHijriah(B)
BM = bulanMasehi(BM)
print("------------------------------------------")
print(f'Tanggal Hijriah : {int(H)} {B} {T} H')
print(f'Tanggal Masehi  : {int(HM)} {BM} {TM} M')
print(f'Hari : {Hari}')
    
    
```

### Konversi Masehi ke Hijriah  (ALGORITMA JEEN - MEUS)
